1Ô∏è‚É£3Ô∏è‚É£ List Comparison

```python
a=[1,2,3,4]
b=[3,4,5,6]

comn_lst=[]
uniq_lst=[]
for i1 in a:
  for j1 in b:
    if i1==j1:
      comn_lst.append(i1)

for i2 in a:
  if i2 not in b:
      uniq_lst.append(i2)

for j2 in b:
  if j2 not in a:
    uniq_lst.append(j2)

print('comman elements in a and b:',comn_lst)
print('unique element in both a and b:',uniq_lst)


```

1Ô∏è‚É£4Ô∏è‚É£ Frequency Without Dict

Count frequency of each element manually.

```python
lst= [1,2,3,4,2,4,2,2,1]
visited=[]
result=[]
cnt=0
for i in lst:
  if i not in visited:
    cnt = 0


    for j in lst:
      if i == j:
        cnt +=1

    result.append([i,cnt])
    visited.append(i)

print(result)
```

1Ô∏è‚É£1Ô∏è‚É£ Pair Sum Problem

```python
ad_lst=[2, 4, 3, 5, 7, 8]
i=0
l_sum=0
pair=[]
sum_lst=[]

while(i<len(ad_lst)):
  pair.append([ad_lst[i],ad_lst[i+1]])


  l_sum = ad_lst[i] + ad_lst[i+1]
  sum_lst.append(l_sum)

  i=i+2

print('pairs :',pair)
print('Pair Sum :',sum_lst)
```
